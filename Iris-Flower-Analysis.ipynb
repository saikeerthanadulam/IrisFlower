SEPERATING INPUT AND OUTPUT COLUMNS

```python
data=df.values
x = data[:,0:4]
y = data[:,4]
print(y)
print(x)
```

SPLITTING DATA INTO TRAINING AND TESTING

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2) 
print(x_test)
```

LOGISTIC REGRESSION 

```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
prediction = model_LR.predict(x_test)
#caluclate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction)*100)
for i in range(len(prediction)):
    print(y_test[i],prediction[i])
```

```python
# a detailed classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction))
```

```python
x_new=np.array([[3,2,1,0.2],[4.9,2.2,3.8,1.1],[5.3,4.5,2.6,1.9]])
prediction1 = model_LR.predict(x_new)
print("Prediction of species: {}".format(prediction1))
```
